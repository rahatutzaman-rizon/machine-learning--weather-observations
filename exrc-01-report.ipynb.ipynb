**2.Calculate the average values**

```python
import csv

def calculate_avg_from_csv(filename):
    total = 0
    count = 0

    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row['ParameterName'].startswith("TA"):
                total += float(row['ParameterValue'])
                count += 1

    if count == 0:
        return 0
    return total / count

# Example usage:
filename = '/content/drive/MyDrive/Weather--/weather_2020-11.csv'
result = calculate_avg_from_csv(filename)
print(f"Average of 'perimeter value' where 'peremeter name' starts with 'TA': {result}")

```

**3.For each parameter, calculate the minimum, maximum, and median values.**

```python
import csv
import numpy as np

def calculate_stats_from_csv(filename):
    values = []

    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row['ParameterName'].startswith("TA"):
                try:
                    value = float(row['ParameterValue'])
                    values.append(value)
                except ValueError:
                    pass  # Skip non-numeric values

    if values:
        min_val = np.min(values)
        max_val = np.max(values)
        median_val = np.median(values)
        return min_val, max_val, median_val
    else:
        return None, None, None

# Example usage:
filename = '/content/drive/MyDrive/Weather--/weather_2020-11.csv'
min_val, max_val, median_val = calculate_stats_from_csv(filename)
if min_val is not None:
    print(f"Minimum value: {min_val}")
    print(f"Maximum value: {max_val}")
    print(f"Median value: {median_val}")
else:
    print("No valid values found for the specified criteria.")
```

**Find the hour with the highest value for each parameter.**

```python
import csv

def calculate_highest_value_from_csv(filename):
    highest_value = None

    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row['ParameterName'].startswith("TA"):
                try:
                    value = float(row['ParameterValue'])
                    if highest_value is None or value > highest_value:
                        highest_value = value
                except ValueError:
                    pass  # Skip non-numeric values

    return highest_value

# Example usage:
filename = '/content/drive/MyDrive/Weather--/weather_2020-11.csv'
highest_value = calculate_highest_value_from_csv(filename)
if highest_value is not None:
    print(f"Highest value: {highest_value}")
else:
    print("No valid values found for the specified criteria.")
```

**4.Create plots to visualize the trends of the following parameters over time**

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_parameter_trends(filename):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(filename)

    # Filter rows where 'ParameterName' starts with 'TA'
    df_ta = df[df['ParameterName'].str.startswith('TA')]

    # Convert 'Time' column to datetime
    df_ta['Time'] = pd.to_datetime(df_ta['Time'])

    # Set 'Time' column as index
    df_ta.set_index('Time', inplace=True)

    # Plot parameter trends over time
    plt.figure(figsize=(10, 6))
    for parameter_name in df_ta['ParameterName'].unique():
        plt.plot(df_ta.loc[df_ta['ParameterName'] == parameter_name].index,
                 df_ta.loc[df_ta['ParameterName'] == parameter_name, 'ParameterValue'],
                 label=parameter_name)
    plt.xlabel('Time')
    plt.ylabel('Parameter Value')
    plt.title('Parameter Trends Over Time')
    plt.legend()
    plt.show()

# Example usage:
filename = '/content/drive/MyDrive/Weather--/weather_2020-11.csv'
plot_parameter_trends(filename)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_parameter_trends(filename):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(filename)

    # Filter rows where 'ParameterName' starts with 'RH'
    df_rh = df[df['ParameterName'].str.startswith('RH')]

    # Convert 'Time' column to datetime
    df_rh['Time'] = pd.to_datetime(df_rh['Time'])

    # Set 'Time' column as index
    df_rh.set_index('Time', inplace=True)

    # Drop rows with NaN values in 'ParameterValue' column
    df_rh.dropna(subset=['ParameterValue'], inplace=True)

    # Plot parameter trends over time
    plt.figure(figsize=(10, 6))
    for parameter_name in df_rh['ParameterName'].unique():
        plt.plot(df_rh.loc[df_rh['ParameterName'] == parameter_name].index,
                 df_rh.loc[df_rh['ParameterName'] == parameter_name, 'ParameterValue'],
                 label=parameter_name)
    plt.xlabel('Time')
    plt.ylabel('Parameter Value')
    plt.title('Parameter Trends Over Time')
    plt.legend()
    plt.show()

# Example usage:
filename = '/content/drive/MyDrive/Weather--/weather_2020-11.csv'
plot_parameter_trends(filename)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_parameter_trends(filename):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(filename)

    # Filter rows where 'ParameterName' starts with 'PRA'
    df_pra = df[df['ParameterName'].str.startswith('PRA')]

    # Convert 'Time' column to datetime
    df_pra['Time'] = pd.to_datetime(df_pra['Time'])

    # Set 'Time' column as index
    df_pra.set_index('Time', inplace=True)

    # Handle NaN values
    df_pra['ParameterValue'] = pd.to_numeric(df_pra['ParameterValue'], errors='coerce')

    # Plot parameter trends over time
    plt.figure(figsize=(10, 6))
    for parameter_name in df_pra['ParameterName'].unique():
        plt.plot(df_pra.loc[df_pra['ParameterName'] == parameter_name].index,
                 df_pra.loc[df_pra['ParameterName'] == parameter_name, 'ParameterValue'],
                 label=parameter_name)
    plt.xlabel('Time')
    plt.ylabel('Parameter Value')
    plt.title('Parameter Trends Over Time')
    plt.legend()
    plt.show()

# Example usage:
filename = '/content/drive/MyDrive/Weather--/weather_2020-11.csv'
plot_parameter_trends(filename)
```
